```python
import pandas as pd
import pydot
from IPython.display import SVG
```

#### Load the example dataset

```python
data_dir = "https://raw.githubusercontent.com/bd2kccd/py-causal/master/data/audiology.txt"
df = pd.read_table(data_dir, sep="\t")
```

#### Load the FGS algorithm from the py-causal library

```python
from pycausal import search as s
```

```python
fgs = s.fgsDiscrete(df,structurePrior = 1.0, samplePrior = 1.0, depth = 3, faithfulness = True, numofthreads = 2, verbose = True, java_max_heap_size = '500M')
```

```python
fgs.getNodes()
```

```python
fgs.getEdges()
```

```python
dot = fgs.getDot()
svg_str = dot.create_svg(prog='dot')
SVG(svg_str)
```
